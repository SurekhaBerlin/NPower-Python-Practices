16. How do you handle missing values in a dataset?

· Drop missing values: df.dropna()

· Fill with a constant: df.fillna(0)

· Fill with mean/median/mode: df['col'].fillna(df['col'].mean())

· Interpolation: df.interpolate()

· Always check df.isna().sum() to see missing counts.

19. What is exploratory data analysis (EDA)?

Answer: EDA is the process of analyzing datasets to summarize main characteristics, often using visualizations and statistics.

· Identify patterns, trends, correlations

· Detect outliers and anomalies

· Understand data distributions

20. What are some common techniques and visualizations used in EDA?

· Summary statistics: df.describe()

· Visualizations:

o Scatterplot → relationships

· Value counts: df['col'].value_counts()

· Groupby aggregations to summarize
