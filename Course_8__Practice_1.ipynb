```python
import pandas as pd
import numpy as np
import matplotlib
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('company_sales_data.csv')
```

```python
df.info()
```

# Exercise 1:
Read Total profit of all months and show it using a line plot

Total profit data provided for each month. Generated line plot must include the following properties:

- X label name = Month Number  
- Y label name = Total profit


```python
# Line plot directly from DataFrame
df.plot(x='month_number', y='total_profit', kind='line')
plt.xlabel('Month Number')
plt.ylabel('Total profit')
plt.show()
```

## Exercise 2:
Get total profit of all months and show line plot with the following Style properties

Generated line plot must include the following style properties:

- Line style dotted and line color should be red  
- Show legend at the lower right location  
- X label name = Month Number  
- Y label name = Sold units number  
- Add a circle marker  
- Line marker color as red  
- Line width should be 3



```python
df.plot(
    x='month_number',                  # column for X-axis
    y='total_profit',                  # column for Y-axis
    linestyle='dotted',
    color='red',
    marker='o',
    markerfacecolor='red',
    linewidth=3,
    label='Total Profit'
)

# Labels
plt.xlabel('Month Number')
plt.ylabel('Sold units number')

# Legend
plt.legend(loc='lower right')

# Display plot
plt.show()

plt.show()
```

# Exercise 3:
Read all product sales data and show it using a multiline plot
Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline
for each product).

```python
df.set_index('month_number', inplace=True)
# Plot all products
df.plot(
    linestyle='solid',  # or 'dotted'
    marker='o',
    linewidth=2
)

plt.xlabel('Month Number')
plt.ylabel('Units Sold')
plt.title('Monthly Units Sold per Product')
plt.legend(title='Products', loc='upper left')
plt.show()
```

## Exercise 4:
 Read toothpaste sales data of each month and show it using a scatter plot. Also, add a grid in the plot. gridline style should .

```python
plt.scatter(df.index, df['toothpaste'], color='blue', label='Toothpaste Sales')

plt.xlabel('Month Number')
plt.ylabel('Units Sold')
plt.title('Monthly Toothpaste Sales')

plt.grid(True, linestyle='--')
plt.legend()
plt.show()
```

## Exercise 5:
Read face cream and facewash product sales data and show it using the bar chart

```python

colors = sns.color_palette("crest", 2)

# Horizontal stacked bar chart
plt.barh(df.index, df['facecream'], color=colors[0], label='Face Cream')
plt.barh(df.index, df['facewash'], color=colors[1], label='Face Wash', left=df['facecream'])

# Labels and title
plt.xlabel('Units Sold')
plt.ylabel('Month Number')
plt.title('Monthly Sales of Face Cream and Facewash')

# Grid with dashed lines
plt.grid(True, linestyle='--', axis='x')

# Legend
plt.legend()

# Show plot
plt.show()
```

## Exercise 6:
Read sales data of bathing soap of all months and show it using a bar chart. Save this plot to
your hard disk

```python

colors = sns.color_palette("crest", 1)

# Bar chart for Bathing Soap
plt.bar(df.index, df['bathingsoap'], color=colors[0])

# Labels and title
plt.xlabel('Month Number')
plt.ylabel('Units Sold')
plt.title('Monthly Sales of Bathing Soap')

# Grid with dashed lines
plt.grid(True, linestyle='--', axis='y')

# Save plot BEFORE showing
plt.savefig('bathing_soap_sales.png')

# Display the plot
plt.show()
```

## Exercise 7:
Read the total profit of each month and show it using the histogram to see the most common
profit ranges

```python
# Histogram
plt.hist(df['total_profit'], bins=5, color=sns.color_palette("crest", 1)[0], edgecolor='black')

# Labels and title
plt.xlabel('Total Profit')
plt.ylabel('Frequency')
plt.title('Histogram of Total Profit per Month')

# Grid with dashed lines
plt.grid(True, linestyle='--', axis='y')

# Show plot
plt.show()
```

## Exercise 8:
Calculate total sale data for last year for each product and show it using a Pie chart


```python
# Calculate total sales for each product
total_sales = df[['facecream','facewash','toothpaste','bathingsoap','shampoo']].sum()

# Pie chart using flare color palette
colors = sns.color_palette("crest", len(total_sales))

plt.pie(total_sales, labels=total_sales.index, autopct='%1.1f%%', startangle=90, colors=colors)
plt.title('Total Sales of Each Product Last Year')
plt.show()
```

## Exercise 9:
Read Bathing soap facewash of all months and display it using the Subplot

```python
colors = sns.color_palette("crest", 2)

# Create subplots (1 row, 2 columns)
fig, axes = plt.subplots(1, 2, figsize=(12,5))

# Bathing Soap plot
axes[0].bar(df.index, df['bathingsoap'], color=colors[0])
axes[0].set_title('Bathing Soap Sales')
axes[0].set_xlabel('Month Number')
axes[0].set_ylabel('Units Sold')
axes[0].grid(True, linestyle='--', axis='y')

# Facewash plot
axes[1].bar(df.index, df['facewash'], color=colors[1])
axes[1].set_title('Face Wash Sales')
axes[1].set_xlabel('Month Number')
axes[1].set_ylabel('Units Sold')
axes[1].grid(True, linestyle='--', axis='y')

# Adjust layout
plt.tight_layout()
plt.show()
```

## Exercise 10:
Read all product sales data and show it using the stack plot

```python
colors = sns.color_palette("crest", len(df.columns)-1)

# Stack plot
plt.stackplot(
    df.index,
    df['facecream'],
    df['facewash'],
    df['toothpaste'],
    df['bathingsoap'],
    df['shampoo'],
    labels=['Face Cream', 'Face Wash', 'Toothpaste', 'Bathing Soap', 'Shampoo'],
    colors=colors
)

# Labels and title
plt.xlabel('Month Number')
plt.ylabel('Units Sold')
plt.title('Stacked Sales of All Products')
plt.legend(loc='upper left')
plt.show()
```
