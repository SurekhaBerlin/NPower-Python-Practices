# 2️⃣ Handle Missing Values

Fill with "Unknown"

```python
df['FIPS'] = df['FIPS'].fillna('Unknown')
df['Admin2'] = df['Admin2'].fillna('Unknown')
df['Province_State'] = df['Province_State'].fillna('Tobefound')
```

Filling with country-level average

```python
df['Lat'] = df['Lat'].fillna(df['Lat'].mean())
df['Long_'] = df['Long_'].fillna(df['Long_'].mean())
```

Fill these columns with O

```python
df['Recovered'] = df['Recovered'].fillna(0)
df['Active'] = df['Active'].fillna(0)
```

```python
df['Incident_Rate'] = df['Incident_Rate'].fillna(0)
```

Small number of missing → fill with 0

```python
df['Case_Fatality_Ratio'] = df['Case_Fatality_Ratio'].fillna(0)
```

Convert Last_Update to datetime (IMPORTANT)

```python
df['Last_Update'] = pd.to_datetime(df['Last_Update'])
```

```python
df.isnull().sum()
```

# 3️⃣ Z-Score Calculation (Recovered column)




```python
from scipy.stats import zscore

# Z-score calculation (ignores NaN automatically)
df['Recovered_Zscore'] = zscore(df['Recovered'], nan_policy='omit')

df[['Recovered', 'Recovered_Zscore']].head()
```

# 4️⃣ Binning (Recovered column)

```python
bins = [-1, 0, 1000, 10000, df['Recovered'].max()]
labels = ['No Recovery', 'Low', 'Medium', 'High']

df['Recovered_Bin'] = pd.cut(df['Recovered'], bins=bins, labels=labels)
```

```python
df.head()
```

# 4️⃣ Country-level aggregation

```python
country_summary = (
    df.groupby('Country_Region')
      .agg(
          Max_Confirmed=('Confirmed', 'max'),
          Max_Deaths=('Deaths', 'max'),
          Max_Recovered=('Recovered', 'max'),
          Max_Active=('Active', 'max')
      )
      .reset_index()
)

country_summary.head()
```

# 6️⃣ Hierarchical Summary (Country & Province)
Goal

Calculate total:

1. Confirmed

2. Deaths

3. Recovered

Grouped by:

1. Country_Region

2. Province_State

```python
df['Recovered'] = df['Recovered'].fillna(0)
df['Province_State'] = df['Province_State'].fillna('Unknown')

hierarchical_summary = (
    df.groupby(['Country_Region', 'Province_State'])
      .agg(
          Total_Confirmed=('Confirmed', 'sum'),
          Total_Deaths=('Deaths', 'sum'),
          Total_Recovered=('Recovered', 'sum')
      )
      .reset_index()
)

hierarchical_summary.head()
```

# 7️⃣ Data Visualization
Line plot for:

1. Confirmed

2. Deaths

3. Recovered

4. Active

Constraint:

✔️ Only countries where total deaths > 150

Step 1: Filter countries by total deaths

```python
# Calculate total deaths per country
deaths_by_country = df.groupby('Country_Region')['Deaths'].sum()

# Filter countries with deaths > 150
high_death_countries = deaths_by_country[deaths_by_country > 150].index

# Filter dataframe
df_filtered = df[df['Country_Region'].isin(high_death_countries)]
```

Step 2: Aggregate by Date & Country

```python
df_filtered = df[df['Country_Region'].isin(high_death_countries)].copy()
df_filtered['Last_Update'] = pd.to_datetime(df_filtered['Last_Update'])
```

```python
df_filtered['Last_Update'] = pd.to_datetime(df_filtered['Last_Update'])

time_series = (
    df_filtered.groupby(['Last_Update', 'Country_Region'])
      .agg(
          Confirmed=('Confirmed', 'sum'),
          Deaths=('Deaths', 'sum'),
          Recovered=('Recovered', 'sum'),
          Active=('Active', 'sum')
      )
      .reset_index()
)
```

Step 3: Line Plot (Matplotlib)

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))

plt.plot(time_series['Last_Update'], time_series['Confirmed'], label='Confirmed')
plt.plot(time_series['Last_Update'], time_series['Deaths'], label='Deaths')
plt.plot(time_series['Last_Update'], time_series['Recovered'], label='Recovered')
plt.plot(time_series['Last_Update'], time_series['Active'], label='Active')

plt.title('COVID-19 Trend (Countries with Deaths > 150)')
plt.xlabel('Date')
plt.ylabel('Number of Cases')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
