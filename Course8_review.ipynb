#**Purpose of Visualization:**

Data visualization is the process of representing data in a visual context (charts, graphs) to make insights easier to understand, spot trends, and communicate findings effectively.

Good visualization helps tell a story with data.

#**Libraries**
**1. Matplotlib**

Definition: Matplotlib is a fundamental Python library for creating static, 2D plots and graphs.

Purpose: It allows full control over plots, including line, scatter, bar, and histogram charts. Great for learning basics of plotting.

**2. Seaborn**

Definition: Seaborn is a statistical data visualization library built on Matplotlib.

Purpose: Makes it easier to create beautiful, informative statistical plots with less code, especially for data analysis.

**3. Plotly**

Definition: Plotly is a Python library for interactive, web-ready visualizations.

Purpose: Allows students to create dynamic charts with hover info, zooming, and filtering, great for dashboards and presentations.

**4. Folium**

Definition: Python library for interactive geospatial maps, built on Leaflet.js.

Purpose: Create maps, markers, and choropleth layers to visualize location-based data.

Choropleth means Instead of placing dots on a map, a choropleth map colors regions (like countries, states, cities, or postal codes) based on a value.

Darker or stronger colors usually mean “more”

Lighter colors usually mean “less”

1. Using Pandas Built-in Plotting

Definition: Pandas has built-in plotting methods that are wrappers around Matplotlib.

Purpose: Quickly create plots directly from DataFrames without writing long Matplotlib code.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample DataFrame
df = pd.DataFrame({
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May'],
    'Sales': [200, 250, 300, 280, 350]
})

# Line plot directly from DataFrame
df.plot(x='Month', y='Sales', marker='o', title='Monthly Sales')
plt.show()


# When you do df.plot(), Pandas internally uses Matplotlib to generate the plot.

# So even though the code looks shorter and simpler, Matplotlib is still the engine behind the scenes.
```

#Choropleth map

A choropleth is a type of map used in data visualization where areas or regions are shaded or colored according to the values of a particular variable.

Key Points:

Purpose

To show how a measurement (numeric data) varies across geographic regions.

Makes patterns and trends visible on a map.

How It Works

Each region (e.g., country, state, county, province) is colored based on the value it represents.

Higher values might use darker shades, lower values lighter shades (or a color gradient).

Examples

Population density by state.

Average income by country.

COVID-19 cases by region.

Libraries to Create Choropleths

Plotly → quick interactive choropleth maps.

Folium → interactive maps with more geographic layers and markers.

```python
import plotly.express as px
import pandas as pd

data = pd.DataFrame({
    'City': ['New York', 'Toronto', 'London'],
    'Lat': [40.7128, 43.651070, 51.5074],
    'Lon': [-74.0060, -79.347015, -0.1278],
    'Value': [100, 200, 300]
})

fig = px.scatter_geo(data, lat='Lat', lon='Lon', text='City', size='Value',
                     projection="natural earth", title="Geospatial Example")
fig.show()
```

#**Creating a Dashboard with Plotly Dash**
1. Introduction

Definition: Dash is a Python framework for building interactive web-based dashboards using Plotly charts.

Purpose: Students can combine multiple visualizations, add filters, dropdowns, and sliders to make data exploration interactive.

| Code Section                    | What it does                                                                 |
| ------------------------------- | ---------------------------------------------------------------------------- |
| `!pip install dash`             | Installs the Dash library (needed for dashboards)                            |
| `app = dash.Dash(__name__)`     | Starts the Dash app                                                          |
| `app.layout = html.Div([...])`  | Defines what the dashboard looks like: title, dropdown, graphs               |
| `dcc.Dropdown(...)`             | Dropdown menu to pick a **day** (`Sun`, `Sat`, etc.)                         |
| `dcc.Graph(id='bar-chart')`     | Placeholder for a bar chart                                                  |
| `dcc.Graph(id='scatter-chart')` | Placeholder for a scatter chart                                              |
| `@app.callback(...)`            | Function that **updates charts** whenever you select a new day               |
| `update_charts(selected_day)`   | Filters the dataset by the selected day and **creates bar + scatter charts** |
| `app.run(debug=True)`           | Runs the dashboard in your browser                                           |


This Dash app is a live, interactive dashboard that lets you explore how tips and bills vary by day and gender, updating charts automatically when you pick a day.

```python
!pip install dash
import dash
from dash import html, dcc
import plotly.express as px
import pandas as pd

# Load sample dataset
df = px.data.tips()

# Initialize the app
app = dash.Dash(__name__)

# Define layout
app.layout = html.Div([
    html.H1("Tips Dashboard"),

    # Dropdown to select day
    dcc.Dropdown(
        id='day-dropdown',
        options=[{'label': day, 'value': day} for day in df['day'].unique()],
        value='Sun'
    ),

    # Graph placeholders
    dcc.Graph(id='bar-chart'),
    dcc.Graph(id='scatter-chart')
])

# Define callback to update charts
from dash.dependencies import Input, Output

@app.callback(
    [Output('bar-chart', 'figure'),
     Output('scatter-chart', 'figure')],
    [Input('day-dropdown', 'value')]
)
def update_charts(selected_day):
    filtered = df[df['day'] == selected_day]

    bar_fig = px.bar(filtered, x='sex', y='total_bill', color='sex', title=f"Total Bill by Gender on {selected_day}")
    scatter_fig = px.scatter(filtered, x='total_bill', y='tip', color='sex', title=f"Tip vs Total Bill on {selected_day}")

    return bar_fig, scatter_fig

# Run the app
if __name__ == '__main__':
  app.run(debug=True,port=8051)




```

| Component        | Purpose                                       |
| ---------------- | --------------------------------------------- |
| `html.Div()`     | Container for HTML elements                   |
| `html.H1()`      | Title/Header                                  |
| `dcc.Graph()`    | Placeholder for Plotly charts                 |
| `dcc.Dropdown()` | Interactive filter                            |
| `dcc.Slider()`   | Another interactive filter option             |
| `@app.callback`  | Function to update charts based on user input |


```python
# Install required packages if not already


!pip install dash
!pip install jupyter-dash
!pip install plotly pandas

from jupyter_dash import JupyterDash
from dash import html, dcc
from dash.dependencies import Input, Output
import plotly.express as px
import pandas as pd



# Load sample dataset
df = px.data.tips()  # Dataset with columns: total_bill, tip, sex, smoker, day, time, size

# Initialize the Dash app
app = dash.Dash(__name__)
app.title = "Tips Dashboard"

# Define layout
app.layout = html.Div([
    html.H1("Interactive Tips Dashboard", style={'textAlign': 'center'}),

    html.Div([
        html.Label("Select Day:"),
        dcc.Dropdown(
            id='day-dropdown',
            options=[{'label': day, 'value': day} for day in df['day'].unique()],
            value='Sun',
            clearable=False
        ),

        html.Label("Select Gender:"),
        dcc.Dropdown(
            id='gender-dropdown',
            options=[{'label': sex, 'value': sex} for sex in df['sex'].unique()],
            value='Male',
            clearable=False
        )
    ], style={'width':'48%', 'display':'inline-block'}),

    html.Div([
        dcc.Graph(id='bar-chart'),
        dcc.Graph(id='scatter-chart')
    ])
])

# Define callback to update charts
@app.callback(
    [Output('bar-chart', 'figure'),
     Output('scatter-chart', 'figure')],
    [Input('day-dropdown', 'value'),
     Input('gender-dropdown', 'value')]
)
def update_charts(selected_day, selected_gender):
    # Filter data
    filtered = df[(df['day'] == selected_day) & (df['sex'] == selected_gender)]

    # Bar chart: Total bill by time
    bar_fig = px.bar(
        filtered,
        x='time',
        y='total_bill',
        color='time',
        title=f"Total Bill by Time ({selected_gender}, {selected_day})"
    )

    # Scatter chart: Tip vs Total Bill
    scatter_fig = px.scatter(
        filtered,
        x='total_bill',
        y='tip',
        size='size',
        color='smoker',
        title=f"Tip vs Total Bill ({selected_gender}, {selected_day})",
        hover_data=['size', 'smoker']
    )

    return bar_fig, scatter_fig

# Run the app
if __name__ == '__main__':
    app.run(debug=True)
```
