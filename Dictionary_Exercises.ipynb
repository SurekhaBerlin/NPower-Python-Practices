```python
#Q1 Create a dictionary for all participants in the JDA program.
#Keys should be the first letter in their names,
#Values should be their names.
#[Hint] dic={"l":(Leul),"J":(Jerry,Jainam)........}

participants = {

    "C": ["Chen", "Christie", "Constance"],
    "E": ["Eliaz"],
    "M": ["Manjot", "Mohammad"],
    "S": ["Surekha"],
    "J" : ["Jerry", "Jainam"]
}
print(participants)
```

```python
#Q2 What are the keys of your dictionary

Participants_Keys = participants.keys()
print("Keys of the dictionary are : ", Participants_Keys)
```

```python
#Q3 What are the values of your dictionary
Participants_Values = participants.values()
print("Values of the dictionary are : ", Participants_Values)
```

```python
#Q4 Find participant's names that start with J,I,M,C,A

for key in participants:
    for name in participants[key]:
        if name.startswith("J") or name.startswith("I") or name.startswith("M") or name.startswith("C") or name.startswith("A"):
            print(name)
```

```python
#Q17 Create the following dictionary
#encryption=
#{10:"hungry",101:"am",110:"I",1011:",",1101:"eat",1001:"want",111:"to"}
# What should be the decryption of the following code
#"110 101 10 1011 110 1001 111 1101"

encryption= {
    10:"hungry",
    101:"am",
    110:"I",
    1011:",",
    1101:"eat",
    1001:"want",
    111:"to"
}
```

```python
# What should be the decryption of the following code
#"110 101 10 1011 110 1001 111 1101"
encoded_message = "110 101 10 1011 110 1001 111 1101"
codes = encoded_message.split()

print(codes)
```

```python
decrypted_words = [encryption[int(code)] for code in codes]

print(decrypted_words)
```

```python
# Join words to form the final message
decrypted_message = " ".join(decrypted_words)

print("Decrypted message:", decrypted_message)
```
