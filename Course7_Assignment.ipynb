```python
# How many men and women (sex feature) are represented in this dataset?
df['gender'].value_counts()
```

```python
df['gender'].unique()
```

```python
# What is the average age (age feature) of women?
average_age_women = df[df['gender'] == ' Female']['age'].mean()
print(average_age_women)
```

```python
# What is the percentage of German citizens (native-country)
percentage_german_citizens = (df[df['native-country'] == ' Germany'].shape[0] / df.shape[0]) * 100
print(percentage_german_citizens)
```

```python
df['income'].unique()
```

```python
# What are the mean and standard deviation of age for those who earn more than 50K per year (salary feature) and those who earn less than 50K per year?

average_age_more_than_50k = df[df['income'] == ' >50K']['age'].mean()
print(average_age_more_than_50k)

average_age_less_than_50k = df[df['income'] == ' <=50K']['age'].mean()
print(average_age_less_than_50k)

```
