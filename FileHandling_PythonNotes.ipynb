**File Handling**

The key function for working with files in Python is the open() function.

The open() function takes two parameters; filename, and mode.

There are four different methods (modes) for opening a file:

"r" - Read - Default value. Opens a file for reading, error if the file does not exist

"a" - Append - Opens a file for appending, creates the file if it does not exist

"w" - Write - Opens a file for writing, creates the file if it does not exist

"x" - Create - Creates the specified file, returns an error if the file exists

**Login**

```python
def sign_up():
    new_name = input("Enter a new username: ")
    new_password = input("Enter a new password: ")

    with open("Credentials.txt", "a") as file:
        file.write(f"{new_name},{new_password}\n")

    print("Account created successfully!")

def login():
    name = input("Enter your username: ")
    password = input("Enter your password: ")

    with open("Credentials.txt", "r") as file:
        credentials = file.readlines()

    for cred in credentials:
        stored_name, stored_password = cred.strip().split(",")
        if name == stored_name:
            if password == stored_password:
                print(f"Welcome {name}!")
                return
            else:
                print("Wrong name or password.")
                return

    print("Your name doesn't exist, please create an account.")

# Create Credentials.txt file with initial data
with open("Credentials.txt", "w") as file:
    file.write("Sarah,12ssr\nLisa,pass\nDonna,D00N\n")

# Test the program
while True:
    choice = input("Do you want to sign up or log in? (sign_up/login/exit): ")
    if choice.lower() == "sign_up":
        sign_up()
    elif choice.lower() == "login":
        login()
    elif choice.lower() == "exit":
        break
    else:
        print("Invalid choice. Please enter 'sign_up', 'login', or 'exit'.")
```
