# Webscraping in 4 steps

Screenshot 2023-10-12 134736.png

```python
# Step 1: Importing libraries
from bs4 import BeautifulSoup
import requests
```

```python

# Step 2: Adding URL
link = "https://www.cricbuzz.com/"
data = requests.get(link)
```

```python
data
```

```python
# Step 3: Creating Soup object to parse data into html format
soup = BeautifulSoup(data.text, "html.parser")
```

```python
soup
```

```python
# Step 4: Retrieve the Data
links = soup.find_all('a',attrs={'class':'text-base wb:text-xl font-bold hover:underline'})
links
```

```python
for i in links:
  print(i.text)
```
