```python
import pandas as pd

# Load the cleaned survey data
df = pd.read_csv("survey_data_updated_new.csv")
```

```python
# --------------------------
# 2️⃣ Clean the data
# --------------------------

# List of important columns to check for missing values
important_cols = [
    'LanguageHaveWorkedWith',
    'DatabaseHaveWorkedWith',
    'PlatformHaveWorkedWith',
    'WebframeHaveWorkedWith',
    'Age'
]

# Drop rows with missing values in any of these columns
df = df.dropna(subset=important_cols)

# Strip leading/trailing whitespace from all skill columns
skill_cols = important_cols[:-1]  # exclude Age
df[skill_cols] = df[skill_cols].apply(lambda x: x.str.strip())
```

```python
# --------------------------
# 3️⃣ Split multi-value columns into separate rows
# --------------------------
def split_column_to_rows(df, column_name, new_name=None):
    """
    Splits a semicolon-separated column into separate rows.

    Parameters:
        df : pandas.DataFrame
            The DataFrame containing the column to split.
        column_name : str
            The name of the column to split.
        new_name : str, optional
            The new name for the split column. If None, defaults to column_name without 'HaveWorkedWith'.

    Returns:
        pandas.DataFrame
            A DataFrame with two columns: 'DeveloperID' and the split values.
    """
    # Determine new column name
    if new_name is None:
        new_name = column_name.replace('HaveWorkedWith','').strip()

    # Split the semicolon-separated values and stack into rows
    split_df = df[column_name].str.split(';', expand=True).stack().reset_index(level=1, drop=True)

    # Reset index and rename columns
    split_df = split_df.reset_index()
    split_df.columns = ['DeveloperID', new_name]

    # Strip extra spaces from values
    split_df[new_name] = split_df[new_name].str.strip()

    return split_df
```

```python
# --------------------------
# Top 10 Programming Languages
# --------------------------
languages_df = split_column_to_rows(df, 'LanguageHaveWorkedWith', new_name='Language')

# Count occurrences and get Top 10
top10_languages = languages_df['Language'].value_counts().head(10).reset_index()
top10_languages.columns = ['Language', 'Count']  # Explicitly rename columns

# Save to CSV
top10_languages.to_csv("top10_languages.csv", index=False)

print(top10_languages.head(10))

```

```python

# --------------------------
# Top 10 Databases
# --------------------------
databases_df = split_column_to_rows(df, 'DatabaseHaveWorkedWith', new_name='Database')

top10_databases = databases_df['Database'].value_counts().head(10).reset_index()
top10_databases.columns = ['Database', 'Count']  # Explicitly rename columns

top10_databases.to_csv("top10_databases.csv", index=False)
print(top10_databases.head(10))
```

```python
# --------------------------
# Top 10 Platforms
# --------------------------
platforms_df = split_column_to_rows(df, 'PlatformHaveWorkedWith', new_name='Platform')

top10_platforms = platforms_df['Platform'].value_counts().head(10).reset_index()
top10_platforms.columns = ['Platform', 'Count']  # Explicitly rename columns

top10_platforms.to_csv("top10_platforms.csv", index=False)
print(top10_platforms.head(10))
```

```python

# --------------------------
# Top 10 Web Frameworks
# --------------------------
frameworks_df = split_column_to_rows(df, 'WebframeHaveWorkedWith', new_name='Framework')

top10_frameworks = frameworks_df['Framework'].value_counts().head(10).reset_index()
top10_frameworks.columns = ['Framework', 'Count']  # Explicitly rename columns

top10_frameworks.to_csv("top10_frameworks.csv", index=False)
print(top10_frameworks.head(10))
```
