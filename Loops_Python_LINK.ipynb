#**Loops in Python**

Loops are used to execute a block of code repeatedly as long as a certain condition is met. Python provides two main types of loops:

  --**for loop**

  --**while loop**



**To create or repeat a loop effectively, four key components are essential:**

**Starting Point (Initialization):** This is where the loop begins. You need to initialize a variable that will act as the starting point for the loop.

**Ending Point (Condition):** This determines when the loop will stop. The loop continues as long as a specific condition is met, and it terminates once this condition becomes false.

**Increment/Decrement (Step or Jump):** This defines how the variable progresses after each iteration. It could be an increase (increment) or a decrease (decrement) in value, which ensures the loop eventually ends.

**Control Variable:** A variable is needed to track the progress of the loop. This variable is updated with the values as the loop executes, moving from the starting point toward the ending point based on the defined step.

#**Common Loop Features in Python: break, continue, and else**

Python provides additional functionality to control the flow of loops through the use of break, continue, and else clauses. These features allow for more precise and flexible control over how a loop behaves during execution.

**break: Exiting the Loop Early**

The break statement is used to terminate the loop immediately, regardless of the loop's condition. Once break is executed, the program exits the loop and continues with the code after the loop.

```python
x=1
while x<=11:
    if x == 5:  # Stop the loop when num is 5
        print("Breaking the loop at:", x)
        break
    print(x)
    x=x+1
```

**continue: Skip the Current Iteration**

The continue statement skips the rest of the code for the current iteration and moves to the next iteration of the loop.

```python
#Skip number 5
num=0
while num<=11:
  num=num+1
  if num==5:
        continue
  else:
     print(num)

```

#**for loop**

A for loop in Python is used to iterate over a sequence (such as a list, tuple, dictionary, string, or range). It executes a block of code repeatedly for each item in the sequence.

#**The range() Function in Python**

The range() function in Python is used to generate a sequence of numbers, which can be iterated over in a loop, typically in a for loop. It provides an easy way to produce sequences of numbers without manually creating a list.

**SYNTAX OF RANGE**

range(start, stop, step)

start (optional): The starting value of the sequence (default is 0).

stop: The number at which the sequence will stop (this value is not included in the sequence).

step (optional): The increment between each value in the sequence (default is 1).

**EXAMPLES**

```python
#FOR LOOP WITH DEFAULT VALUES
for i in range(5):
    print(i)
```

```python
#for loop with step values
for i in range(0, 10, 2):
    print(i)
```

```python
#reversed Sequence
for i in range(10, 0, -2):
    print(i)
```

```python
#sum of a sequence
total = 0
for i in range(1, 6):
    total += i
print(total)
```

#**For loop and Strings**

In Python, you can use a for loop to iterate over the characters in a string. Each character is treated as an element of the string, and the loop processes them one by one.

Here are some examples to demonstrate how for loops can be used with strings:

```python
text = "Python"

for char in text:
    print(char)
```

```python
#If you need to access each character and its index, you can use the enumerate() function to get both the index and character.

text = "Python"

for index, char in enumerate(text):
    print(f"Index {index}: {char}")
```

```python
text = "Python"

# Search for 't' in the string
for char in text:
    if char == 't':
        print("Found 't'!")
        break
else:
    print("'t' not found.")
```

```python
#Counting occurrence of characters

text = "Python is awesome"
char_count = 0

for char in text:
    if char == 'o':
        char_count += 1

print(f"The character 'o' appears {char_count} times.")
```

```python

#Using range with string length

text = "Python"

for i in range(len(text)):
    print(f"Character at index {i}: {text[i]}")
```

```python
#Password Validator

createpass = input("Enter your password: ")

# Initialize flags
has_upper = False
has_lower = False
has_digit = False

# Loop through each character to check conditions
for char in createpass:
    if char.isupper():
        has_upper = True
    if char.islower():
        has_lower = True
    if char.isdigit():
        has_digit = True

# Start validating
valid = True

if len(createpass) < 8:
    print("Password should be at least 8 characters long.")
    valid = False

if not (has_upper and has_lower):
    print("Password should have both capital and small letters.")
    valid = False

if not has_digit:
    print("Password should contain at least one numeric digit.")
    valid = False

if valid:
    print("Password is valid!")
```

#**NESTED LOOPS**


```python
#SQUARE PATTERN

for i in range(5):  # Outer loop for rows
    for j in range(4):  # Inner loop for columns
        print("*", end=" ")  # Print '*' without newline
    print()  # Move to the next line after each row

```

```python
#NUMBER PATTERN

n = 5
for i in range(1, n+1):  # Outer loop for rows
    for j in range(1, i+1):  # Inner loop for columns
        print(j, end=" ")  # Print numbers in each row
    print()  # Move to the next line after each row
```

```python
#DIAMOND SHAPE

n = 5
for i in range(1, n+1):  # Outer loop for the top half of the diamond
    for j in range(n - i):
        print(" ", end="")  # Print spaces for alignment
    for k in range(2 * i - 1):  # Inner loop for stars
        print("*", end="")
    print()

for i in range(n - 1, 0, -1):  # Outer loop for the bottom half of the diamond
    for j in range(n - i):
        print(" ", end="")  # Print spaces for alignment
    for k in range(2 * i - 1):  # Inner loop for stars
        print("*", end="")
    print()
```
