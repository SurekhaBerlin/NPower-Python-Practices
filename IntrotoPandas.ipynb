#Pandas

Pandas is a powerful and open-source Python library designed for data manipulation and analysis. It provides data structures and functions needed to work efficiently with structured data, such as tabular data found in spreadsheets or SQL tables.

**Key Features and Usage**

Pandas is built on top of the NumPy library, which means it leverages many of NumPy's structures and functionalities. It is commonly used in conjunction with other libraries like Matplotlib for plotting, SciPy for statistical analysis, and Scikit-learn for machine learning. Here are some of the key features and uses of Pandas:

**Data Cleaning:** Handling missing data, removing duplicates, and correcting wrong formats.

**Data Manipulation:** Merging, joining, and reshaping datasets.

**Data Analysis:** Grouping data, performing split-apply-combine operations, and calculating statistics.

**Data Visualization:** Plotting data using integrated functions or external libraries like Matplotlib.

Pandas provides three main data structures for handling data: Series, DataFrame, and Panel. A Series is a one-dimensional labeled array that can hold any data type and behaves much like a single column of data, complete with an index. A DataFrame is a two-dimensional table made up of multiple Series, similar to an Excel sheet or SQL table, with rows and columns that can store different data types. Panel, which used to be the three-dimensional data structure in Pandas for handling datasets with multiple layers, has now been deprecated because it was less flexible and less commonly used.

#What is a Series in Pandas?

A Series is basically a single column of data with labels (index).

```python

import pandas as pd

s = pd.Series([10, 20, 30],index=['Mon',"tue","wed"])
print(s)
```

| Feature         | Python List         | NumPy Array                    | Pandas Series         |
| --------------- | ------------------- | ------------------------------ | --------------------- |
| Data type       | Mixed allowed       | Single (uniform)               | Usually uniform       |
| Speed           | Slowest             | Fast                           | Fast                  |
| Math operations |  No                |  Yes (vectorized)             |  Yes (via NumPy)     |
| Index           | Only position-based | Position-based                 | Custom labels + index |
| Best for        | General Python use  | Scientific & numeric computing | Data analysis         |


#Dataframe

A DataFrame in Pandas is a two-dimensional, table-like data structure that looks and behaves very much like an Excel sheet or an SQL table. It consists of rows and columns, where each column is a Pandas Series, and each column can store different data types (numbers, text, dates, booleans, etc.).
A DataFrame also has labels for both rows (index) and columns, which makes filtering, grouping, cleaning, and analyzing data much easier. Because of its flexibility and built-in functions, the DataFrame is the core and most commonly used object in Pandas for data analysis and manipulation.
