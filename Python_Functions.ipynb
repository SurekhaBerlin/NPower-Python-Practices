#**Function with argument and return value**

```python
def square(y):
  area=y*y   #local scope
  return area     # scope of area variable is only limited to square function because it is defined inside it


print("We are solving area of square today")
print()
x=int(input("Enter the side of a square:"))
result=square(x)
print(result)
print("Area of a square is",square(x))

# result=square()
# print("Area of square is",result)
```

```python
#Create a function greet that accepts two arguments: name and greeting, where greeting has a default value of "Hello".
#The function should return a greeting message.

def greet(name, greeting="Hello"):
  message=greeting+" "+name+"!"
  return message


name=input("enter your name: ")
greeting=input("enter a greeting: ")

print(greet(name, greeting))
print(greet(name))
```

```python
#Write a function calculate_discount that takes two arguments â€” price and discount_percentage (default is 10%). The function should return the discounted price

def calculate_discount(price, discount_percentage=10):
  total=price-price*discount_percentage/100
  return total

price=int(input("enter the price: "))
discount_percentage=int(input("enter the discount %: "))

print(calculate_discount(price, discount_percentage))
print(calculate_discount(price))
```

#**Function with multiple return values**

```python
#Write a program to calculate both area and perimeter of a square

def square(side):
  area=side*side
  perimeter=4*side
  return area,perimeter

print(square(5))

a,p=square(6.5)
print("Area is", a,"Perimeter is ",p)
```

```python
def areaofS(a):
  area=a*a
  perimeter=4*a
  return area, perimeter


a,p=areaofS(4)
print("area",a, "Peri",p)
print(areaofS(5))
print("Hello")
```

```python

# Write a function shipping_cost that accepts the following value in argument:

# weight (in kg)
# location (default = "local")

# If location is "local", cost = weight * 5
# If "international", cost = weight * 15
# Return the cost.

def shipping_cost(weight, location="l"):
  if location=="l":
    cost=weight*5
  else:
    cost=weight*15
  TotalCost = cost
  return TotalCost

weight=int(input(" How much does your package weigh(kg)? "))

location=input("Where is the final location? local/international: Enter (L/ I) ")

print(shipping_cost(weight,location.lower()))
print(shipping_cost(weight))
```
